```python
import numpy as np
import tensorflow as tf
import torch
import time

# NumPy FFT Implementation
def fft_numpy(x):
    return np.fft.fft(x)

# TensorFlow FFT Implementation
def fft_tensorflow(x):
    return tf.signal.fft(tf.cast(x, tf.complex64))

# PyTorch FFT Implementation
def fft_pytorch(x):
    return torch.fft.fft(x)

# Example Usage
N = 10000
x_numpy = np.random.rand(N)
x_tensorflow = tf.random.uniform([N])
x_pytorch = torch.rand(N, dtype=torch.complex64)

# NumPy FFT
start_time = time.time()
result_numpy = fft_numpy(x_numpy)
end_time = time.time()
print(f"NumPy FFT Result Size: {len(result_numpy)}")
print(f"NumPy FFT Duration: {end_time - start_time} seconds")

# TensorFlow FFT
start_time = time.time()
result_tensorflow = fft_tensorflow(x_tensorflow)
end_time = time.time()
print(f"TensorFlow FFT Result Size: {len(result_tensorflow)}")
print(f"TensorFlow FFT Duration: {end_time - start_time} seconds")

# PyTorch FFT
start_time = time.time()
result_pytorch = fft_pytorch(x_pytorch)
end_time = time.time()
print(f"PyTorch FFT Result Size: {len(result_pytorch)}")
print(f"PyTorch FFT Duration: {end_time - start_time} seconds")
```

```python
import numpy as np
import tensorflow as tf
import torch
import time

# NumPy Softmax Implementation
def softmax_numpy(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)

# TensorFlow Softmax Implementation
def softmax_tensorflow(x):
    return tf.nn.softmax(x)


def softmax_pytorch(x):
    return torch.nn.functional.softmax(x, dim=0)


N = 10000
x_numpy = np.random.rand(N)
x_tensorflow = tf.random.uniform([N])
x_pytorch = torch.rand(N)


start_time = time.time()
result_numpy = softmax_numpy(x_numpy)
end_time = time.time()
print(f"NumPy Softmax Result Size: {len(result_numpy)}")
print(f"NumPy Softmax Duration: {end_time - start_time} seconds")

# TensorFlow Softmax
start_time = time.time()
result_tensorflow = softmax_tensorflow(x_tensorflow)
end_time = time.time()
print(f"TensorFlow Softmax Result Size: {len(result_tensorflow)}")
print(f"TensorFlow Softmax Duration: {end_time - start_time} seconds")

# PyTorch Softmax
start_time = time.time()
result_pytorch = softmax_pytorch(x_pytorch)
end_time = time.time()
print(f"PyTorch Softmax Result Size: {len(result_pytorch)}")
print(f"PyTorch Softmax Duration: {end_time - start_time} seconds")
```

```python
import numpy as np
import tensorflow as tf
import torch
import time

# NumPy Implementation
def mat_vec_mult_numpy(mat, vec):
    return np.dot(mat, vec)

def vec_add_numpy(vec1, vec2):
    return vec1 + vec2

# TensorFlow Implementation
def mat_vec_mult_tensorflow(mat, vec):
    return tf.linalg.matvec(mat, vec)

def vec_add_tensorflow(vec1, vec2):
    return tf.add(vec1, vec2)

# PyTorch Implementation
def mat_vec_mult_pytorch(mat, vec):
    return torch.mv(mat, vec)

def vec_add_pytorch(vec1, vec2):
    return torch.add(vec1, vec2)

# Example Usage
N = 10000
mat_numpy = np.random.rand(N, N)
vec_numpy = np.random.rand(N)
bias_numpy = np.random.rand(N)

mat_tensorflow = tf.random.uniform([N, N])
vec_tensorflow = tf.random.uniform([N])
bias_tensorflow = tf.random.uniform([N])

mat_pytorch = torch.rand(N, N)
vec_pytorch = torch.rand(N)
bias_pytorch = torch.rand(N)

# NumPy Execution
start_time = time.time()
result_numpy = mat_vec_mult_numpy(mat_numpy, vec_numpy)
result_numpy = vec_add_numpy(result_numpy, bias_numpy)
end_time = time.time()
print(f"NumPy Result Size: {len(result_numpy)}")
print(f"NumPy Duration: {end_time - start_time} seconds")

# TensorFlow Execution
start_time = time.time()
result_tensorflow = mat_vec_mult_tensorflow(mat_tensorflow, vec_tensorflow)
result_tensorflow = vec_add_tensorflow(result_tensorflow, bias_tensorflow)
end_time = time.time()
print(f"TensorFlow Result Size: {len(result_tensorflow)}")
print(f"TensorFlow Duration: {end_time - start_time} seconds")

# PyTorch Execution
start_time = time.time()
result_pytorch = mat_vec_mult_pytorch(mat_pytorch, vec_pytorch)
result_pytorch = vec_add_pytorch(result_pytorch, bias_pytorch)
end_time = time.time()
print(f"PyTorch Result Size: {len(result_pytorch)}")
print(f"PyTorch Duration: {end_time - start_time} seconds")
```
